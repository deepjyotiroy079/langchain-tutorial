### text is spliited based on json values
### chunk size is measured by the number of characters

```python
from langchain_text_splitters import RecursiveJsonSplitter
import json
import requests

response = requests.get('https://jsonplaceholder.typicode.com/posts/').json()

all_docs = []
for item in response:
    json_splitter = RecursiveJsonSplitter(max_chunk_size=100)
    docs = json_splitter.create_documents([item])
    all_docs.append(docs)


for doc in all_docs[:3]:  # Display first 3 items' chunks
    print(doc)
```

```python
# create a document from json chunks

all_texts = []
for item in response:
    json_splitter = RecursiveJsonSplitter(max_chunk_size=100)
    docs = json_splitter.split_text(item)
    all_texts.append(docs)


all_texts[:3]  # Display first 3 items' chunks
```
