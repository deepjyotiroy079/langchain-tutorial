### Splits Texts based on some character sequence
### Chuck size is calculated by number of characters

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('../data/speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=50, chunk_overlap=20, separator="\n")
text_splitter.split_documents(docs)
```

```python
data = ""
with open('../data/speech.txt', 'r') as f:
    data = f.read()

spliiter = CharacterTextSplitter(chunk_size=100, chunk_overlap=50, separator="\n")
text = spliiter.create_documents([data]) # converts normal text to Document type
print(text)
```
